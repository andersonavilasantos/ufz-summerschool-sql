## 1. Why a database (and not a big spreadsheet)?
A single flat file repeats information, is hard to keep consistent, and is slow to query as it grows. A **relational database** splits data into linked **tables**, removes repetition, protects integrity, and lets you *ask questions* with SQL. It is why GenBank, UniProt and SILVA are databases, not spreadsheets.

## 2. The relational model
Data lives in **tables** (also called *relations*). Each **row** is one record (a *tuple*) and each **column** is one attribute (a *field*). One table describes one kind of thing — one table for samples, one for taxa, and so on.

| | genus | phylum |  ← columns (attributes) |
|---|---|---|---|
| **row →** | Pseudomonas | Pseudomonadota | |
| **row →** | Methanosarcina | Euryarchaeota | |

## 3. Keys — how tables connect
- A **primary key (PK)** uniquely identifies each row of a table (`samples.sample_id`, `taxa.taxon_id`, `sites.site`).
- A **foreign key (FK)** is a column that points to another table's primary key. In `abundance`, `sample_id` is a FK to `samples` and `taxon_id` is a FK to `taxa`.

Keys are the whole trick: they let a `JOIN` reconnect the split tables.

## 4. Our schema (an entity–relationship view)
```
sites (site PK)
   │  1
   │        many
samples (sample_id PK, site FK) ──────┐
                                       │ many
                                   abundance (sample_id FK, taxon_id FK, count)
                                       │ many
taxa (taxon_id PK) ────────────────────┘
```
`abundance` is a **junction/fact table**: each row says *how many reads of one taxon were found in one sample*. One sample has many abundance rows; one taxon appears in many samples. That many-to-many link is what we will query.

## 5. What is SQL?
**SQL** (Structured Query Language) is a **declarative** language: you describe *what* you want and the database figures out *how* to get it. It has three families:

| Family | Verbs | Purpose |
|---|---|---|
| **Queries** | `SELECT` | ask questions (our whole focus today) |
| **DML** (data) | `INSERT`, `UPDATE`, `DELETE` | change rows (lesson 09) |
| **DDL** (schema) | `CREATE`, `ALTER`, `DROP` | define tables (lesson 09) |

## 6. Your first query
Run the cell — if you see a number, you are connected.

````python
%%sql
-- how many samples are in the database?
SELECT COUNT(*) AS n_samples FROM samples;
````
